```python
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm
```

```python
for fpath in fm.findSystemFonts(fontpaths=None, fontext='ttf'):
  fm.fontManager.addfont(fpath)

plt.rc("font",family='NanumBarunGothic')
```

### 1) 막대차트(기본)

```python
values = [175,187,184]
labels = ['강백호','서태웅','정대만']

plt.bar(labels,values)
```

><b>색상</b>

```python
values = [175,187,184]
labels = ['강백호','서태웅','정대만']

plt.bar(labels,values,color='r')
```

```python
values = [175,187,184]
labels = ['강백호','서태웅','정대만']

plt.bar(labels,values,color=['r','g','b'])
```

```python
values = [175,187,184]
labels = ['강백호','서태웅','정대만']

plt.bar(labels,values,color=['r','g','b'],alpha=0.3)
```

><b>축 레이블 조정</b>

```python
values = [175,187,184]
labels = ['강백호','서태웅','정대만']

plt.bar(labels,values)
# y축 범위 설정 (최소,최대)
plt.ylim(175,190)
```

><b>막대 너비 조정</b>

```python
values = [175,187,184]
labels = ['강백호','서태웅','정대만']

plt.bar(labels,values,width=0.3)
```

><b>축 레이블 rotation 설정</b>

```python
values = [175,187,184]
labels = ['강백호','서태웅','정대만']

plt.bar(labels,values,width=0.5)
plt.xticks(rotation=45)
plt.yticks(rotation=45)
```

><b>막대 </b>

```python
# x축눈금 값 변경
values = [175,187,184]
labels = ['강백호','서태웅','정대만']
xticks = ['1번학생','2번학생','3번학생']

plt.bar(labels,values)
plt.xticks(labels,xticks)
```

### 2) 막대차트(심화)

**수평막대**

```python
values = [175,187,184]
labels = ['강백호','서태웅','정대만']

plt.barh(labels,values)
```

```python
values = [175,187,184]
labels = ['강백호','서태웅','정대만']

plt.barh(labels,values)
# x축 범위 설정 (최소,최대)
plt.xlim(170,190)
```

**막대 패턴**(https://matplotlib.org/stable/gallery/shapes_and_collections/hatch_style_reference.html)

```python
values = [175,187,184]
labels = ['강백호','서태웅','정대만']

bar = plt.bar(labels,values)
bar[0].set_hatch("/")
bar[1].set_hatch("x")
bar[2].set_hatch("*")
```

```python
values = [175,187,184]
labels = ['강백호','서태웅','정대만']

bar = plt.bar(labels,values)
plt.ylim(170,195)

# rect.get_height() : 막대 높이
for idx, rect in enumerate(bar):
  plt.text(idx, rect.get_height()+0.5, values[idx], ha='center')
```
