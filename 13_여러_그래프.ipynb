### subplots()

```python
# plt.subplots(2,2)
# fig = <Figure size 640x480 with 4 Axes>
# ax = array([[<Axes: >, <Axes: >], [<Axes: >, <Axes: >]], dtype=object)
fig, ax = plt.subplots(2,2, figsize=(15,15))
# 전체 그래프 title
fig.suptitle('여러 그래프 작성')

# 첫번째 그래프 작성(0,0)
ax[0,0].bar(df['이름'],df['국어'],label='국어 점수')
ax[0,0].legend()
ax[0,0].set_title('학생별 국어 점수')
ax[0,0].set_xlabel('이름')

# 두번째 그래프 작성(0,1)
ax[0,1].plot(df['이름'],df['영어'],label='영어 점수')
ax[0,1].plot(df['이름'],df['수학'],label='수학 점수')
ax[0,1].legend()
ax[0,1].set_title('학생별 영어/수학 점수')

# 세번째 그래프 작성(1,0)
ax[1,0].barh(df['이름'],df['키'],label='키')
ax[1,0].legend()
ax[1,0].set_title('학생별 키')

# 네번째 그래프 작성(1,1)
ax[1,1].plot(df['이름'],df['사회'], color='g', alpha=0.5)
```
